Imports

```python
import speech_recognition as sr
from gtts import gTTS
import os
from IPython.display import Audio
```

Functions

```python
# Function 1: to play the correct pronunciation
def play_pronunciation(word):
    tts = gTTS(text=word, lang='hi')
    tts.save("word.mp3")
    os.system("xdg-open word.mp3")  # Command to play audio on Linux
    return Audio("word.mp3", autoplay=True)

# Function 2: to capture the child's speech
def recognize_speech_from_mic():
    recognizer = sr.Recognizer()
    mic = sr.Microphone()

    with mic as source:
        print("Speak now...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        print("Recognizing...")
        # Recognize speech using Google Speech Recognition
        recognized_text = recognizer.recognize_google(audio, language="hi-IN")
        return recognized_text
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand that."
    except sr.RequestError:
        return "Sorry, there seems to be a network issue."

# Function 3: Main function to execute end to end process
def main():
    target_word = "हाँ"  # Change this to any other Hindi word
    print(f"Please say the word: {target_word}")

    # Play pronunciation and display it as an audio widget
    display(play_pronunciation(target_word))

    recognized_word = recognize_speech_from_mic()
    print(f"You said: {recognized_word}")

    if recognized_word == target_word:
        print("Great job! You pronounced it correctly.")
    else:
        print("Hmm, that wasn't quite right. Let's try again.")
```

```python
# Run the main function
main()
```
